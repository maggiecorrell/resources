```python
def get_phonebook():
    phonebook = {}
    
    while True:
        name = input("Enter a name: ")
        
        if name == '':
            break
        
        number = input("Enter a number: ")
    
        phonebook[name] = number
    
    return phonebook
```

```python
phonebook = get_phonebook()
```

```python
phonebook
```

```python
def search_by_name(name, phonebook):
    for user_name, number in phonebook.items():
        if user_name == name:
            return user_name, number
```

```python
search_by_name("Bryce", phonebook)
```

```python
search_by_name("Art", phonebook)
```

```python
def search_by_number(number, phonebook):
    for user_name, user_number in phonebook.items():
        if user_number == number:
            return user_name, number
```

```python
search_by_number('1235', phonebook)
```

```python
search_by_number(1235, phonebook)
```

```python
search_by_number(1337, phonebook)
```

```python
def search(name_or_number, phonebook):
    for user_name, user_number in phonebook.items():
        if user_number == name_or_number or user_name == name_or_number:
            return user_name, user_number
```

```python
search("Sam", phonebook)
```

```python
search('1234', phonebook)
```

```python
search('Art', phonebook)
```

```python
search('12334', phonebook)
```
